松弛法通过迭代更新试探解在整个求解范围（或网格）上的值，直到达到所需的精度，从而解决边值问题。考虑一个常微分方程组 (ODEs) 的边界值问题 （2402.03193）

$$\partial_rF(r)=Y(r;F(r)),$$
定义域为 $(r_a,r_b)$， 边界条件

$$B^{(a)}(r_a;F(r_a))=0,\quad B^{(b)}(r_b;F(r_b))=0,$$

其中 $F, Y$ 以及 $B=(B^{(a)}, B^{(b)})^T$ 是 N 个元素的矢量，如 $F=(F^{(1)},F^{(2)},...F^{(N)})^T$. 现在我们想要通过有限差分的方法求解，我们把空间用网格离散化， $r_{k}= r_{a}+ k( r_{b}- r_{a}) / M$, $k= 0, 1, 2, . . . , M。$ 把 $X$ 在 $r_k$ 的值标记为 $X_k$, 于是ODE和边界条件的有限差分后的形式为

$$E_k(r_k,r_{k-1};F_k,F_{k-1})=0,$$
其中 $k=1,2,...,M$，于是有

$$\begin{aligned}E_k&\equiv F_k-F_{k-1}\\&+(r_{k-1}-r_k)Y\left(\frac{r_{k-1}+r_k}{2},\frac{F_k+F_{k-1}}{2}\right),\end{aligned}$$

我们定义了

$$E_0\equiv(B_0^{(a)},0),\:E_M\equiv(0,B_M^{(b)}),$$

注意 $E_{k}$ 只依赖于 $r_{k},~r_{k-1},~F_{k}$ 和 $F_{k-1}$. 我们可以把不同 k 的 $E_{k}=0$ 表达式组合为 $(M+1)\times N$ 的矩阵方程。
松弛法假设有一个初始的试探解 $P_k=P(r_k)$ 
定义
$$F_k=P_k+\Delta F_k.$$
其中
$\Delta F_{k}=(\Delta F_{k}^{(1)},\Delta F_{k}^{(2)},...,\Delta F_{k}^{(N)})^{T}$ 
对 $E_{k}(P_{k}+\Delta F_{k},P_{k-1}+\Delta F_{k-1})=0$ 做一阶泰勒展开，得到
$$E_{k}(P_{k},P_{k-1})+E_{k,k}\cdot\Delta F_{k}+E_{k,k-1}\cdot\Delta F_{k-1}\approx0,$$
其中 ${\mathrm{defined~}}E_{k,k}=\partial E_{k}/\partial P_{k}{\mathrm{~and~}}E_{k,k-1}=\partial E_{k}/\partial P_{k-1}{\mathrm{~and}}$。 将 $(\Delta F_{k}^{(i)})$ 看作包含指标 $i$ 和 $k$的矢量 $\Delta F$ ，可以很容易得到待求解方程 $\mathcal{A}\Delta F=-E$ 。对这个方程组求解得到 $\Delta F_{k}$, 再将 $F_{k}=P_{k}+\Delta F_{k}$ 作为新的试探解不断迭代，知道满足指定精度。

一般我们可以把精度指定为
$$E_{\mathrm{rel}}=\frac{1}{(M+1)N}\sum_{i=1}^{N}\sum_{k=0}^{M}\left|\frac{\Delta F_{k}^{(i)}}{\max_{k}(F_{k}^{(i)})}\right|,\tag{A8}$$。
如可以选取 $E_{\mathrm{rel}}$ to be around $10^{-15}$。

我们考虑Higgs-portal框架下 Q 球的构型，其运动方程为(arXiv: 2404.16509, Phys. Rev. D 13 (1976) 2739)，

$$
\begin{aligned}
&\partial_r^2 \Phi + \frac{2}{r}\partial_r \Phi + (\nu^2-k^2 H^2)\Phi = 0 \\
&\partial_r^2 H + \frac{2}{r}\partial_r H -k^2H\Phi^2 - \frac{1}{2}H(H^2-1) = 0
\end{aligned}
$$
其边界条件为
$$
\partial_r \Phi\left. \right|_{r= 0} = \partial_r H \left. \right|_{r= 0} = 0, \quad \Phi \left. \right|_{r= \infty} = 0, \quad H \left. \right|_{r= \infty} = 1
$$
定义 $\partial_r \Phi = \Phi_p, \partial_r H = H_p$, 从而将2个2阶微分方程组转化为4个一阶微分方程
$$
\begin{aligned}
&\partial_r \Phi = \Phi_p \\
& \partial_r \Phi_p = -\frac{2}{r}\Phi_p - (\nu^2-k^2 H^2)\Phi \\
& \partial_r H = H_p \\
& \partial H_p = -\frac{2}{r}H_p + k^2 H \Phi^2 + \frac{1}{2}H(H^2-1)
\end{aligned}
$$
有前面可知，在每一个节点k上，error函数的定义为
$$
\begin{aligned}
&E_k^{(0)} = \Phi_k - \Phi_{k-1}-(r_k-r_{k-1}) \frac{\Phi_{p,k}+\Phi_{p,k-1}}{2}\\
&E_k^{(1)} = \Phi_{p,k} - \Phi_{p,k-1} + (r_k-r_{k-1})\left(\frac{2(\Phi_{p,k}+\Phi_{p,k-1})}{r_k+r_{k-1}}+\left[\nu^2-k^2\left(\frac{H_k+H_{k-1}}{2}\right)^2\right]\frac{\Phi_k + \Phi_{k-1}}{2}\right)\\
&E_k^{(2)} = H_k - H_{k-1}-(r_k-r_{k-1}) \frac{H_{p,k}+H_{p,k-1}}{2}\\
&E_k^{(3)} = H_{p,k} - H_{p,k-1} + (r_k-r_{k-1})\left(\frac{2(H_{p,k}+H_{p,k-1})}{r_k+r_{k-1}}-k^2\frac{H_{k}+H_{k-1}}{2}\left(\frac{\Phi_k+\Phi_{k-1}}{2}\right)^2 - \frac{H_k+H_{k-1}}{4}\left[\left(\frac{H_k+H_{k-1}}{2}\right)^2-1 \right]\right)\\
\end{aligned}
$$

将$\Phi, \Phi_p, H, H_p$ 装配成矢量，
$$
y[0, k], y[1,k], y[2,k], y[3,k] = \Phi_k, \Phi_{p,k} , H_k, H_{p,k}
$$
于是有
$$
\begin{aligned}
& \frac{\partial E_k^{(0)}}{\partial y[0,k-1]}=-1 ,\quad \frac{\partial E_k^{(0)}}{\partial y[1,k-1]}=-\frac{r_k-r_{k-1}}{2},\quad\frac{\partial E_k^{(0)}}{\partial y[2,k-1]}=0,\quad\frac{\partial E_k^{(0)}}{\partial y[3,k-1]}= 0\\
&\frac{\partial E_k^{(0)}}{\partial y[0,k]}=1, \quad\frac{\partial E_k^{(0)}}{\partial y[1,k]}=-\frac{r_k-r_{k-1}}{2}, \quad\frac{\partial E_k^{(0)}}{\partial y[2,k]}=0, \quad\frac{\partial E_k^{(0)}}{\partial y[3,k]}=0
\end{aligned}
$$
and
$$
\begin{aligned}
& \frac{\partial E_k^{(1)}}{\partial y[0,k-1]}=\frac{r_k-r_{k-1}}{2}\left[\nu^2-k^2\left(\frac{H_k+H_{k-1}}{2}\right)^2\right], \quad \frac{\partial E_k^{(1)}}{\partial y[1,k-1]}=-1+2\frac{r_k-r_{k-1}}{r_k+r_{k-1}}, \\
&\frac{\partial E_k^{(1)}}{\partial y[2,k-1]}=-\frac{k^2(r_k-r_{k-1})(\Phi_k+\Phi_{k-1})(H_k+H_{k-1})}{4}, \quad \frac{\partial E_k^{(1)}}{\partial y[3,k-1]}= 0\\
&\frac{\partial E_k^{(1)}}{\partial y[0,k]}= \frac{r_k-r_{k-1}}{2}\left[\nu^2-k^2\left(\frac{H_k+H_{k-1}}{2}\right)^2\right], \quad \frac{\partial E_k^{(1)}}{\partial y[1,k]}=1+2\frac{r_k-r_{k-1}}{r_k+r_{k-1}}, \\
&\frac{\partial E_k^{(1)}}{\partial y[2,k]}=-\frac{k^2(r_k-r_{k-1})(\Phi_k+\Phi_{k-1})(H_k+H_{k-1})}{4}, \quad \frac{\partial E_k^{(1)}}{\partial y[3,k]}=0
\end{aligned}
$$
and
$$
\begin{aligned}
&\frac{\partial E_k^{(2)}}{\partial y[0,k-1]}=0, \quad \frac{\partial E_k^{(2)}}{\partial y[1,k-1]}=0, \quad \frac{\partial E_k^{(2)}}{\partial y[2,k-1]}=-1, \quad \frac{\partial E_k^{(2)}}{\partial y[3,k-1]}=-\frac{r_k-r_{k-1}}{2} \\
&\frac{\partial E_k^{(2)}}{\partial y[0,k]}=0, \quad \frac{\partial E_k^{(2)}}{\partial y[1,k]}=0, \quad \frac{\partial E_k^{(2)}}{\partial y[2,k]}=1, \quad \frac{\partial E_k^{(2)}}{\partial y[3,k]}=-\frac{r_k-r_{k-1}}{2}
\end{aligned}
$$
and
$$
\begin{aligned}
& \frac{\partial E_k^{(3)}}{\partial y[0,k-1]}=-\frac{k^2(r_k-r_{k-1})(\Phi_k+\Phi_{k-1})(H_k+H_{k-1})}{4}, \quad \frac{\partial E_k^{(3)}}{\partial y[1,k-1]}=0, \\
&\frac{\partial E_k^{(3)}}{\partial y[2,k-1]}=-\frac{k^2(r_k-r_{k-1})}{2}\left(\frac{\Phi_k+\Phi_{k-1}}{2}\right)^2-\frac{r_k-r_{k-1}}{2}\left[\frac{3}{2}\left(\frac{H_k+H_{k-1}}{2}\right)^2-\frac{1}{2}\right], \quad \frac{\partial E_k^{(3)}}{\partial y[3,k-1]}= -1+2\frac{r_k-r_{k-1}}{r_k+r_{k-1}}\\
&\frac{\partial E_k^{(3)}}{\partial y[0,k]}= -\frac{k^2(r_k-r_{k-1})(\Phi_k+\Phi_{k-1})(H_k+H_{k-1})}{4}, \quad \frac{\partial E_k^{(3)}}{\partial y[1,k]}=0, \\
&\frac{\partial E_k^{(3)}}{\partial y[2,k]}=-\frac{k^2(r_k-r_{k-1})}{2}\left(\frac{\Phi_k+\Phi_{k-1}}{2}\right)^2-\frac{r_k-r_{k-1}}{2}\left[\frac{3}{2}\left(\frac{H_k+H_{k-1}}{2}\right)^2-\frac{1}{2}\right], \quad \frac{\partial E_k^{(3)}}{\partial y[3,k]}=1+2\frac{r_k-r_{k-1}}{r_k+r_{k-1}}
\end{aligned}
$$


为了确定解是否已松弛到目标准确度，我们可以定义以下的相对误差

$$
E_{\mathrm{rel}}=\frac{1}{(M+1) N} \sum_{i=1}^N \sum_{k=0}^M\left|\frac{\Delta F_k^{(i)}}{\max _k\left(F_k^{(i)}\right)}\right|,
$$

当相对误差小于预设的精度，即视作解已经收敛。

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.sparse
import scipy.sparse.linalg
import time

start = time.time() # The time of the beginning
ne = 4 # The number of the ODEs.
mesh = 800 # The number of mesh points.
nl = 2 # The number of boundary conditions of the left.
nr = 2 # The number of boundary conditions of the right. must be nr = ne-nl
itmax = 100 # The maximum number of iterations.
y0_test = np.ones((ne,mesh),dtype=complex)
# An initial guess of y.
x0 = np.linspace(1e-9, 25, mesh)

# 分别定义四个一维数组
#y0_func1 = -0.05 * x0 + 1.0
# 1. 创建一个长度为 mesh 的全零数组
y0_func1 = np.zeros(mesh)
y0_func3 = np.zeros(mesh)

# 2. 计算中间点的索引
# // 确保我们得到一个整数
midpoint_index = mesh // 4 

# 3. 将数组的前半部分（从开始到中间点）赋值为 0.2
y0_func1[:midpoint_index] = 1
y0_func3[midpoint_index:] = 1

y0_func2 = np.full(mesh, 1e-9)
y0_func4 = np.full(mesh, 1e-9)

# 使用 np.vstack 将它们堆叠起来
y0 = np.vstack([
    y0_func1,
    y0_func2,
    y0_func3,
    y0_func4
])

# 转换成复数类型以匹配后续代码
y0 = y0.astype(complex)

# 别忘了将 y 设置为这个新的 y0
y = y0
x = np.linspace(1e-9, 25, mesh) # The meshed points of x.
k0, nu = 3.5, 0.5

print(y0.shape, y0_test.shape)
#print(y0_func3)
```

```python
def difeq(ne,mesh,nl,nr,x,y):
    """
    This function generates the matrix and error column that the relaxation method needs. The purpose is to solve
    A \Delta y = -E
    """
    A = np.zeros((ne*mesh,ne*mesh),dtype=complex)
    E = np.zeros(ne*mesh,dtype=complex)
    for i in range(ne*mesh):
        # Left boundary
        if i < nl:
            A[0,0:ne:1] = np.array([0,1,0,0])
            E[0] = y[1,0]-0
            
            A[1,0:ne:1] = np.array([0,0,0,1])
            E[1] = y[3,0]-0  
            
        # Intermediate points
        elif i>= nl and i < ne*mesh - nr:
            k = (i-nl)//ne + 1 
            h = x[k]-x[k-1]
            
            if (i-nl)%ne == 0:
                A[i,ne*k-ne:ne*k-ne+2*ne:1] = np.array([-1,-0.5*h,0,0,
                                                        1,-0.5*h,0,0])
                E[i] = y[0,k]-y[0,k-1]-0.5*h*(y[1,k]+y[1,k-1])
                
            elif (i-nl)%ne == 1:
                A[i,ne*k-ne:ne*k-ne+2*ne:1] = np.array([0.5*h*(nu**2-k0**2*(y[2,k]+y[2,k-1])**2/4),-1+2*h/(x[k]+x[k-1]),-k0**2*h*(y[0,k]+y[0,k-1])*(y[2,k]+y[2,k-1])/4,0,
                                                        0.5*h*(nu**2-k0**2*(y[2,k]+y[2,k-1])**2/4), 1+2*h/(x[k]+x[k-1]),-k0**2*h*(y[0,k]+y[0,k-1])*(y[2,k]+y[2,k-1])/4,0])
                E[i] = y[1,k]-y[1,k-1]+h*(2/(x[k]+x[k-1])*(y[1,k]+y[1,k-1])+(nu**2-k0**2*(y[2,k]+y[2,k-1])**2/4)*(y[0,k]+y[0,k-1])/2) 
                
            elif (i-nl)%ne == 2:
                A[i,ne*k-ne:ne*k-ne+2*ne:1] = np.array([0,0,-1,-0.5*h,
                                                        0,0,1, -0.5*h])
                E[i] = y[2,k]-y[2,k-1]-0.5*h*(y[3,k]+y[3,k-1])    
                
            else:
                A[i,ne*k-ne:ne*k-ne+2*ne:1] = np.array([-k0**2*h*(y[0,k]+y[0,k-1])*(y[2,k]+y[2,k-1])/4,0,-k0**2*h/2*(y[0,k]+y[0,k-1])**2/4-h/2*(3/2*(y[2,k]+y[2,k-1])**2/4-1/2),-1+2*h/(x[k]+x[k-1]),
                                                        -k0**2*h*(y[0,k]+y[0,k-1])*(y[2,k]+y[2,k-1])/4,0,-k0**2*h/2*(y[0,k]+y[0,k-1])**2/4-h/2*(3/2*(y[2,k]+y[2,k-1])**2/4-1/2),1+2*h/(x[k]+x[k-1])])
                E[i] = y[3,k]-y[3,k-1]+h*(2/(x[k]+x[k-1])*(y[3,k]+y[3,k-1])-k0**2*(y[2,k]+y[2,k-1])/2*(y[0,k]+y[0,k-1])**2/4-1/2*(y[2,k]+y[2,k-1])/2*((y[2,k]+y[2,k-1])**2/4-1))                  
            
        # Right boundary
        else:
            A[-nr,-ne::1] = np.array([1,0,0,0])
            E[-nr] = y[0,-1]-0
            
            A[-nr+1,-ne::1] = np.array([0,0,1,0])
            E[-nr+1] = y[2,-1]-1
    return A, -E
```

```python
slowc = 1000
scalv = np.array([1.0,1.0,1.0,1.0])
conv = 5e-15
scale_adaptive = False
for i in range(itmax):
    A, E = difeq(ne,mesh,nl,nr,x,y)
    A = scipy.sparse.csr_matrix(A)
    Delta_y = scipy.sparse.linalg.spsolve(A, E)
    Delta_y = np.reshape(Delta_y,(ne,mesh),order='F')
    E = np.reshape(E,(ne,mesh),order='F')
    # The typical scale of each component of y is defined
    # as their mean absolute value in the previous iteration.
    # To avoid the division by zero, the initial value of scalv
    # is the lower bound of scalv.
    if scale_adaptive:
        scalv = np.maximum(scalv, np.abs(y).mean(axis=1))
    else:
        pass
    # Original definition of error depends on the delta y,
    # err = np.array([Delta_y[j]/scalv[j] for j in range(len(scalv))])
    # err = np.abs(err).sum()/mesh/ne

    # New definition of error
    err = np.array([E[j]/scalv[j] for j in range(len(scalv))])
    err = np.abs(err).sum()/mesh/ne
    fraction = slowc/10/max(slowc,err)
    if i==0:
        print("       iteration    error")
    print("%8d            %0.6e   "%(i+1,err))
    if err < conv:
        print(' The BVP is solved successfully!')
        break
    elif err >= conv and i <=itmax-1:
        y = y + fraction*Delta_y
        continue
    else:
        print(" Exceed the maximum number of iterations!")
    

end = time.time() # The time of the end.
print("\n The total time of this programme is %0.2e seconds"%(end-start))
```
